## Task 4: Data Cleaning

```python
data.isnull().sum()
```

```python
# Drop missing values
data.dropna(inplace=True)
data.isnull().sum()
```

```python
data.shape
```

```python
# Duplicates values
data.duplicated().sum()

# See the duplicate rows
data[data.duplicated()] 

#Deleting the duplicate rows
data.drop_duplicates(inplace=True)

data.duplicated().sum()
```

```python
data.dtypes
data['id'] = data['id'].astype(object)

data['host_id'] = data['host_id'].astype(object)
```

## Task 5: Data Analysis

**Univariate Anlysis**

```python
# Identifying the outlier in the dataset
df = data[data['price'] < 1500]
sns.boxplot(data=df, x='price')
```

```python
# Price distribution
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='price', bins=100)
plt.title('Price distribution')
plt.ylabel('frequency')
plt.show()
```

```python
df.dtypes
```

```python
#Availability
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='availability_365', bins=100)
plt.title('Availability_365 distribution')
plt.ylabel('frequency')
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean().sort_values(ascending=False)
```

## **Feature Engineering**

```python
# Price per bed
df['Price per bed'] = df['price'] / df['beds']
df.head()
```

```python
df.groupby(by='neighbourhood_group')['Price per bed'].mean().sort_values(ascending=False)
```

## **Bi Variable Analysis**

One Variable deppendcy in another variable

```python
df.columns
```

```python
# Price dependency on the room type
sns.barplot(data=df, x='neighbourhood_group', y='price',hue='room_type')
```

```python
# Number of reviews dependency on the room type
plt.figure(figsize=(8,5))
plt.title('Locality and Review Dependency')
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

```python
df.dtypes
```

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
```

```python
# Geographical distribution of Airbnb Listing
plt.figure(figsize=(10,7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title('Geographical distribution of Airbnb Listing')
plt.show()
```

```python
df.dtypes
```

```python
# Heatmap - correlation of one variable with other for numerical columns
corr = df[['latitude', 'longitude', 'price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
corr
plt.figure(figsize=(8,6))
sns.heatmap(data=corr, annot=True)
```
